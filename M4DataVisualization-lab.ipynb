## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# your code goes here
table_name = 'master'
df = pd.read_sql_query(f"SELECT ConvertedComp FROM master", conn)

plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'], bins=50, edgecolor='k', alpha=0.7)
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df = pd.read_sql_query(f"SELECT Age1stCode FROM master", conn)
df['Age1stCode'] = pd.to_numeric(df['Age1stCode'], errors='coerce')
plt.figure(figsize=(10, 6))
plt.boxplot(df['Age1stCode'], vert=False)
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
df = pd.read_sql_query(f"SELECT Age, WorkWeekHrs FROM {table_name}", conn)
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age and WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
df = pd.read_sql_query(f"SELECT Age, WorkWeekHrs, CodeRevHrs FROM {table_name}", conn)

plt.figure(figsize=(10, 6))
bubble_sizes = df['Age']  # Bubble sizes based on Age column
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=bubble_sizes, alpha=0.5)
plt.title('Bubble Plot of WorkWeekHrs and CodeRevHrs')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)
plt.colorbar(label='Age')
plt.show()
```

Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.

```python
# your code goes here
table_name = 'DatabaseDesireNextYear'
column_name = 'DatabaseDesireNextYear'
df = pd.read_sql_query(f"SELECT {column_name} FROM {table_name}", conn)
#df[column_name] = df[column_name].str.split(';')
#df = df.explode(column_name)
database_counts = df[column_name].value_counts()
database_counts
```

```python
top_5_databases = database_counts.head(5)
labels = top_5_databases.index
sizes = top_5_databases.values
colors = plt.cm.Paired(range(5))

# Create the pie chart
plt.figure(figsize=(10, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.

# Display the pie chart
plt.show()
```


### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

df = pd.read_sql_query(f"SELECT Age, WorkWeekHrs, CodeRevHrs FROM master", conn)
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_data = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median()
plt.figure(figsize=(10, 6))
median_data.plot(kind='bar', stacked=False, alpha=0.7)
plt.title('Median WorkWeekHrs and CodeRevHrs for Ages 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title='Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
df = pd.read_sql_query(f"SELECT Age, ConvertedComp FROM {table_name}", conn)

filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

median_comp = filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(median_comp['Age'], median_comp['ConvertedComp'], marker='o')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation (USD)')
plt.grid(True)
plt.show()
```
